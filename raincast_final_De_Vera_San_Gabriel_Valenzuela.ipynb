## Section 3: Preprocessing

```python
def preprocess_weather_data(df, location_freq=None, scaler=None):
    import pandas as pd
    import numpy as np

    # 1. Data Cleaning
    df = df.copy()

    # Convert date column to datetime
    df['date'] = pd.to_datetime(df['date'])

    # Drop duplicates
    df.drop_duplicates(inplace=True)

    # 2. Feature Engineering

    ## Temporal Features
    df['month'] = df['date'].dt.month
    df['day_of_week'] = df['date'].dt.dayofweek
    df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)

    ## Spatial Features
    # Frequency encoding for location
    if location_freq is None:
        location_freq = df['location'].value_counts(normalize=True)
    df['location_freq'] = df['location'].map(location_freq)

    ## Weather-Based Features
    # Precipitation binning
    df['precip_bin'] = pd.cut(
        df['precipitation'],
        bins=[0, 0.1, 2.0, 20.0],
        labels=["None", "Light", "Moderate/Heavy"],
        include_lowest=True,  # include 0
        right=True
    ).astype(str)

    # Encode precip_bin as ordinal numeric feature
    precip_map = {
        "None": 0,
        "Light": 1,
        "Moderate": 2,
        "Heavy": 3
    }

    df["precip_bin_code"] = df["precip_bin"].map(precip_map).astype(float)
    df['precip_bin_code'] = df['precip_bin_code'].fillna(0)

    # Pressure drop (lag feature)
    df.sort_values(by=['location', 'date'], inplace=True)
    df['pressure_lag1'] = df.groupby('location')['pressure'].shift(1)
    df['pressure_drop'] = df['pressure_lag1'] - df['pressure']
    df['pressure_drop'] = df['pressure_drop'].fillna(0)

    ## Interaction Features
    df['humid_precip'] = df['humidity'] * df['precipitation']

    ## Drop date column
    df = df.drop(columns=["date", "location", "precip_bin", "pressure_lag1"])

    # 3. Standardization of numerical features
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    # Exclude target column if present
    for col in ["rain_tomorrow", "month", "day_of_week", "is_weekend", "precip_bin_code"]:
        if col in numeric_cols:
            numeric_cols.remove(col)

    if scaler is None:
        scaler = StandardScaler()
        df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
    else:
        df[numeric_cols] = scaler.transform(df[numeric_cols])

    print(numeric_cols)

    return df
```

```python
def set_seeds(seed=42):
    """Set seeds for reproducibility"""
    np.random.seed(seed)
    random.seed(seed)
    os.environ["PYTHONHASHSEED"] = str(seed)
```

```python
def load_data(dataset_path, num_samples=None, seed=1234):
    ds = ray.data.read_csv(dataset_path)
    ds = ds.random_shuffle(seed=seed)

    # Optional sampling
    if num_samples:
        sampled = ds.take(num_samples)
        ds = ray.data.from_items(sampled)

    # Apply prepare_data preprocessing
    ds = ds.map_batches(prepare_data, batch_format="pandas")

    return ds
```

```python
class RainfallPreprocessor:
    def _fit(self, dataset: ray.data.Dataset):
        df = dataset.to_pandas()
        self.location_freq = df['location'].value_counts(normalize=True).to_dict()

    def _transform(self, dataset: ray.data.Dataset) -> ray.data.Dataset:
        return dataset.map_batches(
            lambda df: preprocess_weather_data(df, location_freq=self.location_freq),
            batch_format="pandas"
        )

    def _fit_transform(self, dataset: ray.data.Dataset) -> ray.data.Dataset:
        self._fit(dataset)
        return self._transform(dataset)
```

```python
def select_estimator_class(model_type):
    if model_type == "rf":
        return RandomForestClassifier
    elif model_type == "logreg":
        return LogisticRegression
    else:
        raise ValueError(f"Unknown model_type: {model_type}")
```

```python
def load_model_from_checkpoint(checkpoint):
    path = checkpoint.to_directory()
    import joblib
    return joblib.load(f"{path}/model.pkl")
```

```python
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    roc_auc_score
)

def evaluate_model(model, X_val, y_val, model_name="RainCast", show_plot=True):
    """
    Evaluate a classification model on validation data.

    Parameters:
    - model: Trained classifier with .predict() and optionally .predict_proba()
    - X_val: Validation features
    - y_val: Validation labels
    - model_name: Optional name for display
    - show_plot: Whether to display the confusion matrix heatmap

    Returns:
    - Dictionary with accuracy, ROC-AUC (if available), and confusion matrix
    """
    y_pred = model.predict(X_val)

    # Try to get predicted probabilities for ROC-AUC
    y_proba = None
    if hasattr(model, "predict_proba"):
        try:
            y_proba = model.predict_proba(X_val)[:, 1]
        except (IndexError, AttributeError):
            pass

    # Accuracy
    acc = accuracy_score(y_val, y_pred)
    print(f"\n--- {model_name} Evaluation ---")
    print(f"Accuracy: {acc:.4f}")

    # Classification report
    print("\nClassification Report:")
    print(classification_report(y_val, y_pred, digits=4))

    # ROC-AUC
    auc = None
    if y_proba is not None and len(np.unique(y_val)) == 2:
        auc = roc_auc_score(y_val, y_proba)
        print(f"ROC-AUC: {auc:.4f}")

    # Confusion matrix
    cm = confusion_matrix(y_val, y_pred)
    if show_plot:
        plt.figure(figsize=(5, 4))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
        plt.title(f"{model_name} - Confusion Matrix")
        plt.xlabel("Predicted Label")
        plt.ylabel("True Label")
        plt.tight_layout()
        plt.show()

    return {
        "accuracy": acc,
        "roc_auc": auc,
        "confusion_matrix": cm
    }
```

### 1.3 Tune Configuration

RunConfig and Checkpoint were also previously configured.

```python
# RandomForest Hyperparameters
search_space_rf = {
    "model_type": "rf",
    "params": {
        "n_estimators": tune.choice([100, 200, 300, 400, 500]),
        "max_depth": tune.choice([5, 10, 15, None]),
        "min_samples_split": tune.choice([2, 5, 10]),
        "min_samples_leaf": tune.choice([1, 2, 4])
    }
}
```

```python
# LogisticRegression Hyperparameters
search_space_logreg = {
    "model_type": "logreg",
    "params": {
        "C": tune.loguniform(0.01, 10),
        "penalty": tune.choice(["l1", "l2"]),
        "solver": tune.choice(["liblinear", "saga"])
    }
}
```

```python
def run_tuning(model_type, train_ds, val_ds, scaling_config, run_config, label_column="rain_tomorrow", num_samples=20):
    # Pick search space based on model type
    if model_type == "rf":
        search_space = search_space_rf
    elif model_type == "logreg":
        search_space = search_space_logreg
    else:
        raise ValueError("model_type must be 'rf' or 'logreg'")

    search_alg = OptunaSearch()
    scheduler = ASHAScheduler()

    # Wrapper function: build trainer, fit, return metrics
    def train_fn(config):
        trainer = build_trainer(config, train_ds, val_ds,
                            label_column, scaling_config, run_config)
        result = trainer.fit()
        metrics = result.metrics
        print("Trial metrics:", metrics)  # Debugging

        # Extract validation score
        val_score = metrics.get("val", {}).get("test_score")
        if val_score is None:
            raise ValueError(f"No validation score found in {metrics}")

        # Report it as "accuracy" so Tune can optimize it
        train.report({"accuracy": float(val_score)})

    tuner = tune.Tuner(
        train_fn,
        param_space=search_space,
        tune_config=tune.TuneConfig(
            search_alg=search_alg,
            scheduler=scheduler,
            num_samples=num_samples,
            metric="accuracy",   # must match the reported key
            mode="max"
        ),
        run_config=run_config
    )

    results = tuner.fit()
    best_result = results.get_best_result(metric="accuracy", mode="max")

    print(f"Best {model_type} config:", best_result.config)
    print(f"Best {model_type} metrics:", best_result.metrics)

    return results, best_result
```

### Behavioral Testing

```python
def behavioral_tests(model, X_sample, y_sample):
    results = {}

    # --- Invariant Test ---
    # Slight perturbation in temperature should not change prediction
    X_invariant = X_sample.copy()
    X_invariant["temperature"] = X_invariant["temperature"] + 0.5
    pred_original = model.predict(X_sample)[0]
    pred_invariant = model.predict(X_invariant)[0]
    results["Invariant"] = (pred_original == pred_invariant)

    # --- Directional Expectation Test ---
    # Increase humidity and precipitation should increase probability of rain
    X_directional = X_sample.copy()
    X_directional["humidity"] = X_directional["humidity"] + 20
    X_directional["precipitation"] = X_directional["precipitation"] + 2
    proba_original = model.predict_proba(X_sample)[0][1]
    proba_directional = model.predict_proba(X_directional)[0][1]
    results["Directional"] = (proba_directional > proba_original)

    # --- Minimum Functionality Test ---
    # Clear rain scenario: high humidity, high precipitation, low pressure
    X_minfunc = X_sample.copy()
    X_minfunc["humidity"] = 95
    X_minfunc["precipitation"] = 10
    X_minfunc["pressure"] = 980
    pred_minfunc = model.predict(X_minfunc)[0]
    results["Minimum Functionality"] = (pred_minfunc == 1)  # Expect rain

    return results

# Example usage with one test sample
sample = X_test.iloc[[0]].copy()  # pick one test row
behavioral_results = behavioral_tests(trained_model, sample, y_test.iloc[[0]])
print(behavioral_results)
```

### Batch Inference

```python
# Load new batch of weather data
new_data = pd.read_csv("new_weather_data.csv")
new_data = prepare_data(new_data)
```

```python
class RainCastPredictor:
    def __init__(self, model, scaler=None, location_freq=None):
        """
        Initialize the Predictor with a trained model, optional scaler, and location frequency map.
        """
        self.model = model
        self.location_freq = location_freq

    def preprocess(self, df):
        """
        Use the existing preprocess_weather_data function.
        """
        # Call your existing preprocessing function
        processed = preprocess_weather_data(
            df,
            location_freq=self.location_freq
        )
        return processed

    def predict_batch(self, df):
        """
        Run batch predictions and return dataframe with results.
        """
        processed = self.preprocess(df)

        # Drop target column if present
        if "rain_tomorrow" in processed.columns:
            processed = processed.drop(columns=["rain_tomorrow"])

        preds = self.model.predict(processed)
        probas = self.model.predict_proba(processed)

        df = df.copy()
        df["rain_prediction"] = preds
        df["rain_probability"] = probas[:, 1]
        return df

    def predict_single(self, row):
        """
        Predict for a single row (dict or Series).
        """
        df = pd.DataFrame([row])
        processed = self.preprocess(df)

        if "rain_tomorrow" in processed.columns:
            processed = processed.drop(columns=["rain_tomorrow"])

        pred = self.model.predict(processed)[0]
        proba = self.model.predict_proba(processed)[0][1]
        return {"rain_prediction": int(pred), "rain_probability": float(proba)}
```

```python
# Initialize predictor with trained model
predictor = RainCastPredictor(trained_model)

# Batch inference
results = predictor.predict_batch(new_data)
print(results.head())
```
